```python
import torch
torch.__version__
```

Tensors

```python
#salar - tensor with dim(zero)
tensor0 = torch.tensor(1)
tensor0.ndim
```

```python
#vector - tensor with dim(1)
tensor1 = torch.tensor([1,2,3,4,5])
print(f"Vector: {tensor1} , Dim: {tensor1.ndim}, Shape: {tensor1.shape}")
#matrix - tensor with dim(2)
tensor2 = torch.tensor([[1,0,0],[0,1,0]])
print(f"Matrix: {tensor2} , Dim: {tensor2.dim()}, Shape: {tensor2.size()}")
#ndim = dim and shape = size
```

```python
#Ways to create tensors
#using Size
size = (3,4)
tensor4 = torch.empty(size)
#Creating Random Tensor (values auto asigned btwn. 0 and 1)
tensor5 = torch.rand(size)
tensor5.dtype
# .dtype returns the tensor type (int,float et. al.) Default type: float32
# type can be be assigned during the creation or can be modified later using type(torch.newType)
```

```python
#Creating Tensors using Numpy Array
import numpy as np

exampleArray = np.array([[1,2],[3,4]])
tensor6 = torch.tensor(exampleArray) #Passed by Value
tensor7 = torch.from_numpy(exampleArray) #Share Same memory - Passed by Reference

exampleArray *= 5
print(exampleArray)
print(tensor6) # No UPDATION
print(tensor7) 
```

Device is the attribute of the tensor that tells the location of the tensor (loc on GPU or CPU)

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
tensor8 = torch.ones(3,7).to(device) #"CPU" -> "cuda"
tensor9 = torch.zeros(3,7, device = device) # -> 'cuda' 
```

Assessing Elements in Tensor

```python
print(tensor7[0])
print(tensor7[1,0])

#Slicing
print(tensor7[:, 1]) #Column
print(tensor7[0,:]) #Row
```

Operation of Tensor

```python
tensor10 = torch.rand(2,3)
tensor11 = torch.ones(2,3)
print(tensor10)
print(tensor11)
```

```python
# elementwise addition
tensor12 = tensor10 + tensor11
# torch.add(tensor10,tensor11)
print(tensor12)

# elementwise subtraction
tensor13 = tensor10 - tensor11
# torch.sub(tensor10,tensor11)
print(tensor13)

# elementwise multiplication
tensor14 = tensor10 * tensor11
# torch.mul(tensor10,tensor11)
print(tensor14)

# elementwise division
tensor15 = tensor10 / tensor11
# torch.div(tensor10,tensor11)
print(tensor15)
```
