**PANDAS**

```python
#Reading Data
import pandas as pd

df = pd.read_csv('datasets\\flavors_of_cocoa.csv',
                 index_col=0 , #Use Column 0 as index
                 na_values = ['!', 'NA', 'NaN', '--']  #Treats these values as NaN 
                 )
print('Done')
```

```python
#Dataframe Inspection
print(df.head(2))
print(df.tail(2))
```

```python
df.info()
```

```python
df.describe()
```

```python
#Attributes of the Dataframe
print(df.shape)
print(df.dtypes)
print(df.columns)
```

```python
#Accessor Methods
print(df.loc[10, "Bean Origin"])
print(df.iloc[10,1])
print(df.at[10, "Bean Origin"])
print(df.iat[10,1])
```

```python
#Memory Optimisation 

print(df['Bean Origin'].memory_usage(deep=True))

#Optimising
df['Bean Origin'] = df['Bean Origin'].astype('category')
print(df['Bean Origin'].memory_usage(deep=True))
```

```python
print(df["Rating"].mean())
print(df["Rating"].median())
print(df["Bean Origin"].mode())
```

```python
pd.crosstab(index=df['Bean Origin'], 
            columns=df['Rating'],normalize= 'index')
```
